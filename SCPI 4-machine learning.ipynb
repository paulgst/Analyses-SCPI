### Préparation des données pour le machine learning

On va à présent importer le dataframe que l'on a sauvegardé, puis créer des jeux de données différents ainsi que faire les prétraitements nécessaires (split, normalisation standard, etc...)

```python
import numpy as np
import pandas as pd

import matplotlib.pyplot as plt
%matplotlib inline

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler, StandardScaler
from sklearn.metrics import mean_absolute_error, mean_squared_error, mean_absolute_error, r2_score

df = pd.read_excel('df.xlsx',index_col=0)
df
```

```python
X = df.drop(['scpi_liste','pct_distrib_2022','variation_prix_2022'],axis = 1)
y = df.pct_distrib_2022
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.2)

sc_fit = StandardScaler().fit(X_train)
X_train_sc = sc_fit.transform(X_train)
X_test_sc = sc_fit.transform(X_test)
```

```python
sc_fit_all = StandardScaler()
X_sc = sc_fit_all.fit_transform(X)
```

### Machine learning avec toutes les données

On va effectuer une cross-validation en essayant plusieurs modèles de machine learning. On va d'abord importer les librairies nécessaires :

```python
from sklearn.linear_model import LinearRegression, LassoCV, RidgeCV
from sklearn.neighbors import KNeighborsRegressor
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import cross_validate, cross_val_score
```

```python
md = LinearRegression()
cv = cross_validate(md, X_sc, y, scoring=['neg_mean_squared_error'], cv=3, n_jobs=5, verbose=2)
print(f"score LinearRegression : {cv['test_neg_mean_squared_error'].mean()}")

md = LassoCV()
cv = cross_validate(md, X_sc, y, scoring=['neg_mean_squared_error'], cv=3, n_jobs=5, verbose=2)
print(f"score LassoCV : {cv['test_neg_mean_squared_error'].mean()}")

md = RidgeCV()
cv = cross_validate(md, X_sc, y, scoring=['neg_mean_squared_error'], cv=3, n_jobs=5, verbose=2)
print(f"score RidgeCV : {cv['test_neg_mean_squared_error'].mean()}")
```

```python
md = KNeighborsRegressor()
cv = cross_validate(md, X, y, scoring=['neg_mean_squared_error'], cv=3, n_jobs=5, verbose=2)
print(f"score : {cv['test_neg_mean_squared_error'].mean()}")
```

```python
md = RandomForestRegressor()
cv = cross_validate(md, X, y, scoring=['neg_mean_squared_error'], cv=3, n_jobs=5, verbose=2)
print(f"score : {cv['test_neg_mean_squared_error'].mean()}")
```

### Avec un nouveau jeu de donnée, cible selon les 2 années précédentes, de l'année, de la catégorie et du type

Selon les analyses faites dans le chap. 4, on déduit que le pourcentage de distribution dépends principalement :
- de l'année et son contexte
- des pourcentages de distribution précédents
- du type de scpi

on va alors décomposer le jeu de données avec une variable cible, dépendant de l'année (en tant que variable continue) du pourcentage 1 et 2 années avant et du type de SCPI.

```python
df_new1 = df[['scpi_liste', 'pct_distrib_2020', 'pct_distrib_2021',
       'pct_distrib_2022', 'categorie_Activités / Logistique',
       'categorie_Bureaux', 'categorie_Commerce', 'categorie_Diversifiée',
       'categorie_Résidentiel', 'categorie_Santé et éducation',
       'categorie_Tourisme et loisirs']]#'pct_distrib_2018','pct_distrib_2019',
df_new1['annee']='2022'
df_new1 = df_new1[['scpi_liste', 'annee','pct_distrib_2020', 'pct_distrib_2021', 'categorie_Activités / Logistique',
       'categorie_Bureaux', 'categorie_Commerce', 'categorie_Diversifiée',
       'categorie_Résidentiel', 'categorie_Santé et éducation',
       'categorie_Tourisme et loisirs','pct_distrib_2022']]

df_new1 = df_new1.rename(columns={"pct_distrib_2020": "pct_distrib_N-2","pct_distrib_2021": "pct_distrib_N-1","pct_distrib_2022": "pct_distrib"})


df_new2 = df[['scpi_liste', 'pct_distrib_2019', 'pct_distrib_2020',
       'pct_distrib_2021', 'categorie_Activités / Logistique',
       'categorie_Bureaux', 'categorie_Commerce', 'categorie_Diversifiée',
       'categorie_Résidentiel', 'categorie_Santé et éducation',
       'categorie_Tourisme et loisirs']]
df_new2['annee']='2021'
df_new2 = df_new2[['scpi_liste', 'annee','pct_distrib_2019', 'pct_distrib_2020', 'categorie_Activités / Logistique',
       'categorie_Bureaux', 'categorie_Commerce', 'categorie_Diversifiée',
       'categorie_Résidentiel', 'categorie_Santé et éducation',
       'categorie_Tourisme et loisirs','pct_distrib_2021']]

df_new2 = df_new2.rename(columns={"pct_distrib_2019": "pct_distrib_N-2","pct_distrib_2020": "pct_distrib_N-1","pct_distrib_2021": "pct_distrib"})


df_new3 = df[['scpi_liste', 'pct_distrib_2018', 'pct_distrib_2019',
       'pct_distrib_2020', 'categorie_Activités / Logistique',
       'categorie_Bureaux', 'categorie_Commerce', 'categorie_Diversifiée',
       'categorie_Résidentiel', 'categorie_Santé et éducation',
       'categorie_Tourisme et loisirs']]
df_new3['annee']='2020'
df_new3 = df_new3[['scpi_liste', 'annee','pct_distrib_2018', 'pct_distrib_2019', 'categorie_Activités / Logistique',
       'categorie_Bureaux', 'categorie_Commerce', 'categorie_Diversifiée',
       'categorie_Résidentiel', 'categorie_Santé et éducation',
       'categorie_Tourisme et loisirs','pct_distrib_2020']]

df_new3 = df_new3.rename(columns={"pct_distrib_2018": "pct_distrib_N-2","pct_distrib_2019": "pct_distrib_N-1","pct_distrib_2020": "pct_distrib"})

df_new=pd.concat([df_new1,df_new2,df_new3],axis=0)
df_new
```

```python
df_new= df_new.astype({'annee' : 'float64'})
df_new_ml=df_new.drop('scpi_liste',axis=1)
df_new_ml.info()

X_new_ml = df_new_ml.drop(['pct_distrib'],axis = 1)
y_new_ml = df_new_ml.pct_distrib
```

```python
sc_fit_all = StandardScaler()
X_new_ml_sc = sc_fit_all.fit_transform(X_new_ml)
```

```python
md = LinearRegression()
cv = cross_validate(md, X_new_ml_sc, y_new_ml, scoring=['neg_mean_squared_error'], cv=3, n_jobs=5, verbose=2)
print(f"score LinearRegression : {cv['test_neg_mean_squared_error'].mean()}")

md = LassoCV()
cv = cross_validate(md, X_new_ml_sc, y_new_ml, scoring=['neg_mean_squared_error'], cv=3, n_jobs=5, verbose=2)
print(f"score LassoCV : {cv['test_neg_mean_squared_error'].mean()}")

md = RidgeCV()
cv = cross_validate(md, X_new_ml_sc, y_new_ml, scoring=['neg_mean_squared_error'], cv=3, n_jobs=5, verbose=2)
print(f"score RidgeCV : {cv['test_neg_mean_squared_error'].mean()}")
```

L'essai n'est pas concluant car on a une moyenne du carrée des erreurs plus importante qu'en appliquant les mêmes modèles sur les jeux normaux.

En conclusion, parmis les différents tests, le modèle de régression Lasso avec toutes les données est le meilleur.
